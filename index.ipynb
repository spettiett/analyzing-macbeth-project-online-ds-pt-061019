- Read in the Macbeth text

- Create a dictionary to track the {key:value} pair for each word {word : word count}
- Loop through each "word" using the dic.get(w,0)+1   --- Returns the value for key if found in dict
  -- (if word is found, then increment the word count in the dict for that word)

```python
# Read in the Macbeth text file
```

```python
import requests
macbeth = requests.get('http://www.gutenberg.org/cache/epub/2264/pg2264.txt').text

print(type(macbeth))
print(len(macbeth))
print(macbeth[:500])
```

```python
# Split the transcript into words
```

```python
convert_to_words = macbeth.lower().split()
```

```python
convert_to_words
```

```python
# Create a dictionary
```

```python
dic = {}
```

```python
# Iterate through the text of Macbeth
```

```python
# Update word counts

for w in convert_to_words:
    dic[w] = dic.get(w,0)+1   # Returns the value for key if found in dict
print ('\n'.join(['%s,%s' % (k, v) for k, v in dic.items()]))
```

```python
type(dic.items())
```

```python
dic.items()
```

```python
# Sort the word counts in order

from collections import OrderedDict
from operator import itemgetter

ordered = OrderedDict(sorted(dic.items(), key=itemgetter(1), reverse=True))

toptwentyfive = dict(list(ordered.items())[:25])
print(toptwentyfive) 
```

```python
x=toptwentyfive.keys()
x
```

```python
y=toptwentyfive.values()
y
```

```python
# Create Bar Graph
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.bar(x,y, align='center')
plt.show()
```

```python
# Include descriptive titles and labels
```

```python
plt.figure(figsize=(20,10))

plt.bar(x,y, align='center', color="g")


plt.title("25 Most Common Words in Shakespeare's Macbeth", fontsize=14)
plt.xlabel("Word", fontsize=10)
plt.ylabel("Word Count", fontsize=10)

plt.show()
```
